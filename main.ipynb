```python
image = Image.open('gradient.jpg') # Загружаем изображение используя библиотеку Image
width = image.size[0]
height = image.size[1]
pix = image.load() # Создаем ссылку на объект PixelAccess
image
```

# Image Topology (A x B)

```python
pix_list = []    # Лист для значений пикселей
a = 0    # Длинная сторона картинки

if width > height:
    a = width
elif width < height:
    a = height
    
for i in range(a):
    li = []
    for j in range(a):
        try:
            li.append(pix[i,j][0])
        except:
            li.append(0)
    pix_list.append(li)

pix_ar = np.array(pix_list)

pix_ar
```

```python
x = [i for i in range(0, a)]
y = [i for i in range(0, a)]

X, Y = np.meshgrid(x, y)
#z = f(X,Y,pixel_list)
z = pix_ar

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, cmap='Spectral')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
